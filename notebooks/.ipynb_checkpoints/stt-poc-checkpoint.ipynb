```python
import sys
import os

PROJECT_ROOT = os.path.abspath("..")
sys.path.append(PROJECT_ROOT)

PROJECT_ROOT
```

```python
from src.stt.stt_pipeline import run_stt_pipeline

print("Imports OK.")
```

```python
video_path = "../data/sample.mp4"   

chunks, json_path = run_stt_pipeline(
    video_file=video_path,
    model_size="tiny",     # or "small" if faster
    device="cpu"             # change to "cuda" if using GPU
)

print("Transcript saved to:", json_path)
print("Number of segments:", len(chunks))

# Display first few segments
for c in chunks[:5]:
    print(f"[{c['start']:.2f} - {c['end']:.2f}] {c['text']}")
```

```python
import json

with open(json_path, "r", encoding="utf-8") as f:
    chunks_loaded = json.load(f)

print("Loaded segments:", len(chunks_loaded))
chunks_loaded[:3]
```

```python
readable_transcript = "\n".join(
    f"[{c['start']:.2f}] {c['text']}"
    for c in chunks_loaded
)

print(readable_transcript[:1500])
```

```python
import matplotlib.pyplot as plt

times = [c['start'] for c in chunks_loaded]
plt.plot(times)
plt.title("Segment Start Times")
plt.xlabel("Segment Index")
plt.ylabel("Start (sec)")
plt.show()
```
