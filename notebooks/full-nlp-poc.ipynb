```python
json_path = PROJECT_ROOT / "transcripts" / "sample.json"

with open(json_path, "r", encoding="utf-8") as f:
    whisper_segments = json.load(f)

print(f"Loaded transcript with {len(whisper_segments)} segments.\n")

# ------------------------------------
# RUN FULL NLP PIPELINE
# ------------------------------------

results = run_full_nlp_pipeline(whisper_segments)

chunks = results["chunks"]
topics = results["topics"]
chapters = results["chapters"]
boundaries = results["boundaries"]

print("â”€â”€â”€â”€â”€â”€â”€â”€ NLP Pipeline Overview â”€â”€â”€â”€â”€â”€â”€â”€")
print(f"ğŸ§© Merged Chunks:       {len(chunks)}")
print(f"ğŸ“˜ Topics Detected:     {len(topics)}")
print(f"ğŸ“Œ Valid Boundaries:    {results['boundaries']}")
print(f"ğŸ“ Raw Boundaries:      {results['raw_boundaries']}")
print(f"ğŸ“– Chapter Summaries:   {len(chapters)}")
print("\n")


# ============================================
# PRETTY PRINT CHAPTERS
# ============================================

for chap in chapters:
    print("=" * 100)
    print(f"ğŸ“Œ Topic {chap['topic_id'] + 1}")
    print(f"â±  {chap['start']} â†’ {chap['end']}\n")

    print("ğŸ¬ Title:")
    print(chap.get("title", "(missing)"))
    print("\n")

    print("ğŸ“ Summary:")
    print(chap.get("summary", "(missing summary)"))
    print("\n")

    print("ğŸ”¹ Bullet Points:")
    for b in chap.get("bullets", []):
        print(f" - {b}")
    print("\n")

    print("ğŸ”‘ Keywords:")
    print(", ".join(chap.get("keywords", [])))
    print("\n")
```
